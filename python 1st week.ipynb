```python
def is_palindrome(lst):
    copied_list = lst.copy()
    copied_list.reverse()

    if lst == copied_list:
        return True
    else:
        return False

list1 = [1, 2, 3, 2, 1]
list2 = [1, "abc", "abc", 1]

print("List1:", list1)
print("Is Palindrome?", is_palindrome(list1))

print("\nList2:", list2)
print("Is Palindrome?", is_palindrome(list2))
```
